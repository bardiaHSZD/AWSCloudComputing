```python
a = np.array([0,5,7,3])
print(type(a))
print(a.shape)

#inclusive
print(a[2:])
#exclusive
print(a[:2]) 

print(a[[-1]])
print(a[-1])

print(a[::2])

# Sum of all elements
print(a.sum())
# Standard Deviation
print(a.std())
# Variance
print(a.var())
# Mean of all elements
print(a.mean())
# Minimum and Maximum
print(a.min(), a.max())

b = np.array([0.1, 2.0, 3.4])
print(b.dtype)

a_matrix = np.array([[1,2,3],
                    [3.2,4.3,1]])
print(a_matrix.shape)

# Sum of all elements
print(a_matrix.sum())
# Standard Deviation
print(a_matrix.std())
# Variance
print(a_matrix.var())
# Mean of all elements
print(a_matrix.mean())
# Minimum and Maximum
print(a_matrix.min(), a.max())
# summation of columns
print(a_matrix.sum(axis=0))
# summation of rows
print(a_matrix.sum(axis=1))
# Minimum and Maximum of Columns
print(a_matrix.min(axis=0), a_matrix.max(axis=0))
# Minimum and Maximum of Rows
print(a_matrix.min(axis=1), a_matrix.max(axis=1))
# Mean of Columns and Rows
print(a_matrix.mean(axis=0), a_matrix.mean(axis=1))

b_matrix = np.array([
    [
        [1,2],
        [3,3],
        [8,8]
    ],
    [
        [5,6],
        [7,7],
        [9,9]
    ]
])

print(b_matrix.shape)
```

```python
import numpy as np

a = np.array([
    [0 , 1, 2],
    [3, 4, 5]
])

b = np.array([
    [6, 7, 8],
    [9, 10, 11]
])

# dot product
print('The dot product a*b: \n')
print(a*b)

# transpose
print('The transpose pf b: \n')
print(b.T)

# matrix product
print('The matrix product of a and b: \n')
print(a @ (b.T))
```

```python
import numpy as np

print(np.arange(10))
print(np.arange(5,10))
print(np.arange(1,10,2))
print(np.arange(1,10,0.5))

print(np.arange(10).reshape(5,2))
print(np.random.randint(1,10, (10)))
print(np.random.randint(1,10, (5,2)))

print(np.random.rand(5,2))
# normal distribution random float generator (center, std, shape)
print(np.random.normal(0, 0.1, (5,2)))
# uniform distribution random float generator (min, max, shepe)
print(np.random.uniform(0, 1, (5,2)))
# uniform nardom sorted set of floats
print(np.sort(np.random.uniform(0, 1, (10))))


print(np.zeros((4)), np.ones((4)))
print(np.zeros((2,2)), '\n' ,np.ones((2,2)))
print(np.ones((3,2), dtype=int))
```

# **Week 3: Data visualization - Matplotlib**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.arange(10)
y = np.arange(0, 1000, 100)

plt.plot(x, y)
```

```python
x = np.random.normal(0.5, 0.1, (100))
x.sort()
y = x **2
print(x)
print(y)
plt.plot(x, y)
```

```python
plt.plot(x, y, 'o')
```

```python
x = np.arange(100)
y = x**3
plt.plot(y, '1')
```

```python
markers = ['o', '*', '.', ',', 'x', 'X', '+', 'P', 's', 'D', 'd', 'p', 'H', 'h', 'v', '^', '<', '>', '1', '2', '3', '4', '|', '_']

for m in markers:
  print(m)
  plt.plot(y, m)
  plt.show()
```

```python
y = np.array([1, 3, 13])
line = ['-',':','--','-.']
for l in line:
  print(l)
  l = 'o' + l + 'r'
  plt.plot(y, l)
  plt.show()
```

```python
color = ['r','g','b','c','m','y','k','w']
for c in color:
  print(c)
  c = 'o:' + c
  plt.plot(y, c)
  plt.show()
```

```python
plt.plot(y, marker = '^', ms = 30)
```

```python
x = np.arange(10)
y1 = x*2
y2 = x**2
plt.plot(y1)
plt.plot(y2)
```

```python
plt.plot(x, y1, label = 'double')
plt.plot(x, y2, label = 'square')
plt.xlabel("Xs")
plt.ylabel("Ys")
plt.grid()
plt.legend(['1st', '2nd'])
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y1, label = 'double')
plt.legend()
plt.xlabel("Xs")
plt.ylabel("Ys")
plt.subplot(2, 1, 2)
plt.plot(x, y2, label = 'square')
plt.legend()
plt.xlabel("Xs")
plt.ylabel("Ys")
plt.grid()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y1, label = 'double')
plt.legend()
plt.xlabel("Xs")
plt.ylabel("Ys")
plt.subplot(1, 2, 2)
plt.plot(x, y2, label = 'square')
plt.legend()
plt.xlabel("Xs")
plt.ylabel("Ys")
plt.grid()
```
